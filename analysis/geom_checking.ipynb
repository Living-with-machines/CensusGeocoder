```python
import pandas as pd
import geopandas as gpd
import pathlib
```

```python

list_of_census = [
    "EW_1851",
    "EW_1861",
    "EW_1881",
    "EW_1891",
    "EW_1901",
    "EW_1911",
    "scot_1851",
    "scot_1861",
    "scot_1871",
    "scot_1881",
    "scot_1891",
    "scot_1901",
]

list_of_census = ["EW_1911"]

for geom in ["os_open_roads"]:

    for census in list_of_census:
        country = census.split("_")[0]
        year = census.split("_")[1]

        linked_dir = pathlib.Path(f"output/{year}_old/{country}/{geom}/linked")

        gis_file = gpd.read_file(linked_dir.parent / f"{geom}_{country}_{year}.geojson")
        # print(gis_file)

        # all_df = pd.DataFrame()
        df_list = []
        for linked_file in linked_dir.iterdir():
            df = pd.read_csv(linked_file, sep="\t", usecols= ["unique_add_id", f"os_open_roads_{country}_{year}"])
            df_list.append(df)
        all_df = pd.concat(df_list)
all_df
```
