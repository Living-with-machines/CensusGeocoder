```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
geocode = pd.read_csv("analysis_alignment_test3/1911/EW/link_lkup_merged_geoms.tsv", sep = "\t", usecols = ["RecID", "os_open_roads_EW_1911"])
# geocode = pd.read_csv("analysis_alignment_test3/EW_all_years.tsv")
```

```python
geocode = geocode.dropna()
geocode
```

```python
census = pd.read_csv("analysis_alignment_test3/EW_all_years.tsv", sep = "\t", usecols = ["RecID", "Occode", "year", "ParID"])
census = census[census["year"] == 1911].copy()
```

```python
census
```

```python
geocode_census = pd.merge(left = census, right = geocode, on = "RecID", how = "right")
geocode_census
```

```python
bikes = geocode_census[geocode_census["Occode"] == 362]
bikes_grouped = bikes.groupby("os_open_roads_EW_1911").size().reset_index()
bikes_grouped
```

```python
open_roads_data = "output_geom_test/1911/EW/os_open_roads/os_open_roads_EW_1911.geojson"
open_roads = gpd.read_file(open_roads_data)
```

```python
cov_roads = open_roads[open_roads["conparid_01-11"] == 106064].copy()
```

```python
cov_roads.head()
```

```python
cov_roads_bikes = pd.merge(left = bikes_grouped, right = cov_roads, on = "os_open_roads_EW_1911", how = "inner")
cov_roads_bikes
```

```python
cov_roads_bikes.to_csv("cov_road_bikes.tsv", sep = "\t", index = False)
```

```python
list_of_cols = ['ParID','TOWN']
urban_classification = pd.read_excel('England_Wales_Urban_Classification_1851_1911.xlsx',sheet_name='1911',usecols=list_of_cols)

urban_classification_dict = dict(zip(urban_classification['ParID'],urban_classification["TOWN"]))
```

```python
geocode_census
```

```python
geocode_census["ParID"] = pd.to_numeric(geocode_census["ParID"], downcast="integer")
```

```python
geocode_census['TOWN'] = geocode_census['ParID'].map(urban_classification_dict)
```

```python
geocode_census
```

```python
geocode_census_sample = geocode_census[geocode_census["TOWN"].str.contains("BIRMINGHAM|COVENTRY|NOTTINGHAM|LONDON|WOLVERHAMPTON|MANCHESTER|LIVERPOOL", na=False)].copy()
geocode_census_sample
```

```python
geocode_census_sample["bike"] = np.where(geocode_census_sample["Occode"] == 362, 1,0)
geocode_census_sample.head()
```

```python
geocode_census_sample["street_total"] = geocode_census_sample.groupby("os_open_roads_EW_1911")["bike"].transform("sum")
geocode_census_sample
```

```python
m_graph = geocode_census_sample.drop_duplicates(subset="os_open_roads_EW_1911").copy()
m_graph
```
